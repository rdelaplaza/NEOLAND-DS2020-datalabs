### TEST de memoría

```python
# Cambiamos el tipo de dato
import numpy as np
diamonds_f32 = pd.read_csv("diamonds_original.csv",
                        nrows=1000,
                        dtype={
                            "carat" : np.float32,
                            "depth" : np.float32,
                            "table" : np.float32,
                            "x" : np.float32,
                            "y" : np.float32,
                            "z" : np.float32,
                            "price" : np.int16
                            })
```

```python
diamonds_f32.info()
```

Con el solo cambiar el tipo de dato de `float64` a `float32` hemos ahorrado algo de memoria.

```python
# Procedemos con el cambio de categorícas
diamonds_cat = pd.read_csv("diamonds_original.csv",
                        nrows=1000,
                        dtype={
                            "carat" : np.float32,
                            "depth" : np.float32,
                            "table" : np.float32,
                            "x" : np.float32,
                            "y" : np.float32,
                            "z" : np.float32,
                            "price" : np.int16,
                            "cut" : "category",
                            "color" :"category",
                            "clarity":"category"
                            })
```

```python
diamonds_cat.info()
```

Otra forma de cargar los datos y optimizar la memoria y los tiempos de carga.

```python
cols = [
    "carat",
    "cut",
    "color",
    "clarity",
    "depth",
    "table",
    "price"
]
```

```python
# Utilizando el parámetro usecols cargamos solo el array declarado anteriormente con las columnas interesadas 
diamonds4 = pd.read_csv("diamonds_original.csv",
                        nrows=1000,
                        dtype={
                            "carat" : np.float32,
                            "depth" : np.float32,
                            "table" : np.float32,
                            "x" : np.float32,
                            "y" : np.float32,
                            "z" : np.float32,
                            "price" : np.int16,
                            "cut" : "category",
                            "color" :"category",
                            "clarity":"category"
                            },
                       usecols=cols)
```

```python
diamonds4.info()
```

```python
# Utilizamos el paráemtro chuncksize para realizar cargas de n elementos
diamonds_5 = pd.read_csv("diamonds_original.csv",
                        nrows=1000,
                        dtype={
                            "carat" : np.float32,
                            "depth" : np.float32,
                            "table" : np.float32,
                            "x" : np.float32,
                            "y" : np.float32,
                            "z" : np.float32,
                            "price" : np.int16,
                            "cut" : "category",
                            "color" :"category",
                            "clarity":"category"
                            },
                       usecols=cols,
                        chunksize=200)
```

```python
diamonds_5
```

```python
diamonds_5.info()
```

Para poder leer este formato TextFileReader necesitaremos iterar sobre el objeto generado.

```python
def process(df):
    return f"processed {df.size} items"
for chunk in diamonds_5:
    process(chunk)
```

```python
chunk
```

```python
chunk.info()
```
