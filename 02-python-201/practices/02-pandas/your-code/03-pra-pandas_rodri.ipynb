### Ejercicio 2

Agrupa los datos cargados en el ejercicio 1 por **año** y, para cada año, muestra el número total de películas (*Movies*) y series de TV (*TV Show*) por separado. 

**Nota:** Al Notebook de teoría hemos visto cómo calcular la media de una agrupación de datos. Para resolver este ejercicio necesitarás investigar cómo contar (*count*) todos los valores de una columna en una agrupación. Te puede ser de utilidad la función [aggregate](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.aggregate.html) de pandas, la que nos permite aplicar diferentes funciones a una columna en un ```groupby```.


````python
# Respuesta
# Observamos la parte que nos interesa del dataframe

df_netflix[['release_year', 'type']]
````

````python
# Agrupamos por año y por tipo (serie o película)

df_netflix_groupedby_year = df_netflix.groupby(['release_year', 'type'])


# Utilizamos la función 'aggregate' para contar los elementos por los que hemos agrupado

df_netflix_groupedby_year.aggregate('count')
````

````python
# Utilizo un for...loop para mostrar los resultados de series y películas reproducidas cada año

for name,group in df_netflix_groupedby_year:
    if name[1] == 'TV Show':
        print ('En',name[0],'se reprodujeron',len(group), 'series.')
    if name[1] == 'Movie':
        print ('En',name[0],'se reprodujeron',len(group), 'peliculas.')
````

### Ejercicio 3

Muestra las películas estrenadas el _2019_ que tienen una duración superior a _100_ minutos. 

**Nota:** Al Notebook de teoría hemos visto operaciones básicas de filtrado. Para resolver este ejercicio necesitarás investigar cómo aplicar condiciones más complejas. 

**Nota 2:** Los valores de la columna de duración (*duration*) son strings, ya que combinan números y letras. Por lo tanto, se tienen que transformar a int una vez se haya construido el subset de datos de películas. Te puede ser de utilidad la función [astype](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.astype.html) de pandas.
 


````python
# Respuesta
# Dividimos la cadena del campo 'duration' por el primer espacio que aparezca, entre el número (duración de la película o temporadas de la serie) y la cadena ('min' or 'Season')

df_netflix.duration= df_netflix.duration.str.partition(' ')[0]

df_netflix
````

````python
# Convertimos la columna 'duration' en 'integer' para poder comprarla con el número 100

df_netflix.duration = df_netflix.duration.astype(int)
````

````python
# Creamos el subset solamente con los datos que nos interesan y observamos los datos

df_netflix_peliculas = df_netflix[['type', 'title', 'release_year', 'duration']]

df_netflix_peliculas
````

````python
# Filtramos por tipo (película), año de estreno (2019) y duración (+100 min)

df_netflix_peliculas_2019 = df_netflix_peliculas[(df_netflix_peliculas.type == 'Movie') & (df_netflix_peliculas.release_year == 2019) & (df_netflix_peliculas.duration > 100)]

df_netflix_peliculas_2019
````

### Ejercicio 5

Crea una copia del dataframe original añadiendo una nueva columna que nos indicará por cada fila la duración de la película. Definimos tres niveles:

- **short** : si la duración es inferior a 90 minutos. 
- **medium** : si la duración está entre 90 y 150 minutos.
- **long** : si la duración es superior a 150 minutos.

Muestra por pantalla aquellas filas que correspondan a una duración corta ( _short_ ). 

**Nota:** pandas nos ofrece principalmente tres maneras de añadir nuevas columnas a un dataframe existente. Cualquiera de las tres formas son válidas para resolver este ejercicio. Podéis buscar en Internet cómo añadir columnas a un dataframe de pandas, ya que la documentación presenta los tres métodos de forma separada.


````python
# Respuesta
# Creamos un nuevo dataframe solamente con las películas

new_df_netflix = df_netflix.drop(df_netflix[df_netflix.type == 'TV Show'].index)

new_df_netflix
````

````python
# Creamos una función que asigne rangos en función de la duración de las películas

def ranges(movie):
    movie = movie.duration
    if movie < 90:
        return 'short'
    elif movie > 150:
        return 'long'
    else:
        return 'medium'
````

````python
# Añadimos una nueva columna al dataframe para los difrentes rangos de duración de las películas, aplicando a la columna la función creada

new_df_netflix['duration_range'] = new_df_netflix.apply(ranges, axis = 1)

new_df_netflix
````

````python
# Mostramos solamente aquellas que corresponden al rango de duración 'short'

new_df_netflix_short = new_df_netflix[new_df_netflix.duration_range == 'short']
new_df_netflix_short
````

En este ejercicio trabajaremos con el dataset de Netflix, pero analizando las series de TV. Muestra un gráfico de barras con el número de series de TV americanas agrupadas por categoría (*listed_in*).


Consideraciones:

- El eje de las Y debe incluir las etiquetas de las categorías de las series de TV.
- Debe incluir un texto con el número de series por cada categoría del gráfico. El texto debería estar situado a la derecha de cada barra.

**Nota**: En el Notebook de teoría hemos visto cómo mostrar gráficos lineares simples con el matplotlib. Para este ejercicio explora las funcionalidades que presenta la librería pandas para generar gráficos a partir de dataframes. También pueden ser de utilidad las funciones [text](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.text.html) o [annotate](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.annotate.html) para mostrar el número de series al lado de cada barra. 


````python
# Respuesta
# Importamos la librería 'matplotlib' donde hemos importado el resto de librerías
# Esta vez vamos a trabajar solamente con las series

df_netflix_series = df_netflix.drop(df_netflix[df_netflix.type == 'Movie'].index)

df_netflix_series
````

````python
# Como, además, vamos a representar solamente las series americanas, simplificamos aún más nuestro dataframe

df_netflix_series_americanas = df_netflix_series.drop(df_netflix_series[df_netflix_series.country != 'United States'].index)

df_netflix_series_americanas
````

````python
# Independizamos las categorías y contamos los valores ('value_counts')

df_netflix_categorias_lista = df_netflix_series_americanas.listed_in.str.split(',').explode()

type(df_netflix_categorias_lista)

df_netflix_categorias_lista.value_counts()
````

````python
# Debido a los espacios, existen valores que se contabilizan separados aunque en realidad son el mismo. Eliminamos estos espacios para consolidar nuestros datos

df_netflix_categorias_lista_consolidada = df_netflix_categorias_lista.str.strip()

df_netflix_categorias_lista_consolidada.value_counts()
````

````python
# Como la información que queremos visualizar mediante el diagrama de barras es un objeto 'Series', lo vamos a convertir en un dataframe

type(df_netflix_categorias_lista_consolidada.value_counts())

serie = df_netflix_categorias_lista_consolidada.value_counts()

df_categorias = serie.to_frame()

df_categorias
````

````python
# Mostramos los datos de manera horizontal (barras horizontales)

grafico_barras_horizontales = df_categorias.plot.barh(y = 'listed_in', color = 'red', figsize = (20, 10))
````
