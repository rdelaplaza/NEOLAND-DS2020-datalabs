```python
# Importamos las librerías
import random

# Declaramos las variables
value = random.randint(1,6)
count = 0
guess = 0

# Creamos el iterador
while guess != value:
    count += 1 # Incremento
    guess = input("Adivina un número entre 1 y 6: ")
    if guess.isnumeric():
        guess = int(guess)
else:
    print('Has adivinado en {} ocasiones!'.format(count))

    
```

```python
## MEJORA DEL JUEGO ANTERIOR

# Importamos las librerías
import random

# Declaramos las variables
value = random.randint(1,10)
count = 0
guess = 0

# Creamos el iterador
while guess != value:
    count += 1 # Incremento
    guess = input("Adivina un número entre 1 y 10: ")
    if guess.isnumeric():
        guess = int(guess)
        print('Este es tu intento número {}'.format(count))
    else:
        print('Por favor, introduzca un número entero.')
        continue
    if guess > value:
        print('El número es demasiado alto, vuelva a intentarlo')
    if guess < value:
        print('El número es demasiado bajo,  vuelva a intentarlo')
else:
    print('Has adivinado en {} ocasiones!'.format(count))
```
